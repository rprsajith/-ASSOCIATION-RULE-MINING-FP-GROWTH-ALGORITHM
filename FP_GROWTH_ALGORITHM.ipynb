```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
data = [['VOLLEYBALL','VOLLEYBALLNET','FOOTBALL','CRICKET','BASKETBALL','BATMITTON'],
        ['CRICKET BAT','FOOTBALLSHOE','KABADDI','VOLLEYBALL','FOOTBALL','CRICKET'],
        ['BATMITTONBAT','BASKETBALL','KABADDISHOE','CRICKET','VOLLEYBALL','BATMITTON'],
        ['VOLLEYBALLNET','BASKETBALLSHOE','VOLLEYBALL','CRICKETBAT','FOOTBALL','BATMITTONBAT'],
        ['CRICKET','KABADDI','VOLLEYBALLNET','FOOTBALLSHOE','KABADDISHOE','BATMITTON']]
```

```python
te = TransactionEncoder()
te_ary = te.fit_transform(data)
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
fpgrowth(df,min_support=0.6)
```

```python
fpgrowth(df,min_support=0.6,use_colnames=True)
```
